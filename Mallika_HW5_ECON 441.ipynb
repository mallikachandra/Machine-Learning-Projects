# 3.) Clean your data for a decision tree


```python
X = pd.get_dummies(df_select, "class_of_admission")
y = df["case_status"]
```

```python
df_select = df_select.dropna()
```

```python
# Select the columns you want to create dummies for
columns = ["class_of_admission", "country_of_citizenship", "job_info_work_state"]

# Create dummies for the selected columns
a = pd.get_dummies(df_select[columns])
```

```python
# Concatenate the original DataFrame with the dummy variables
X = pd.concat([df_select, a], axis=1)

# Drop the original columns that were used to create the dummies
X.drop(columns, axis=1, inplace=True)
X = X.drop(['case_status'], axis = 1)
y = df_select['case_status']
```

# 4.) Fit and plot a decision tree of depth X


```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf.fit(X,y)
```

```python
plt.figure(figsize = (20,8))
tree.plot_tree(clf,feature_names = X.columns )
```

# 6.) Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
from sklearn.model_selection import train_test_split
max_depths = [4,5,6]
outputs = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

for md in max_depths:
  clf = tree.DecisionTreeClassifier(max_depth = md)
  clf.fit(X_train,y_train)

  oos_accuracy =  clf.score(X_test, y_test)
  outputs.append(oos_accuracy)
    
```
