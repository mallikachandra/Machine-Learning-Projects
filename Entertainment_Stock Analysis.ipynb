## Explore Stocks in the Entertainment and Streaming Services Space

such as Netflix, Inc. (NFLX), The Walt Disney Company (DIS), Roku, Inc. (ROKU), and Spotify Technology S.A. (SPOT).

```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['NFLX', 'DIS', 'ROKU', 'SPOT']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)
```

```python
df = pd.concat(df_list, keys = tickers, names = ['Ticker', 'Date'])
```

```python
df.reset_index(inplace=True)
# Convert the 'Date' column to datetime type for correct plotting
df['Date'] = pd.to_datetime(df['Date'])
```

```python
import plotly.express as px
```

```python
fig = px.line(df, x = 'Date',
             y = 'Close',
             color = 'Ticker',
             title = 'Stock Market Analysis of Major Players in the Entertainment Space')
fig.show()
```
